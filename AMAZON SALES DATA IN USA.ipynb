# Fix The Missing Values

```python
missing_values=['Postal_Code']
for i in missing_values:
    if data[i].dtype=='object':
        data[i].fillna(data[i].mode()[0],inplace=True)
    else:
        data[i].fillna(round(data[i].mean()),inplace=True)
```

```python
data.isnull().sum()
```

```python
ca=[]
nu=[]
for i in data.columns:
    if data[i].nunique()<20:
        ca.append(i)
    else:
        nu.append(i)
cat=data.loc[:,ca]
num=data.drop(ca,axis=1)
```

# CATEGORICAL COLUMNS

```python
for i in cat.columns:
    if data[i].isnull().sum()==0:
        print(f'{i}')
        print(f"Missing Values :{data[i].isnull().sum()}")
        print(f"Mode value :{data[i].mode()}")
        print(f"Unique value :{data[i].unique()}")        
        print(f"No of Unique value :{data[i].nunique()}")
        print('\n')
```

# NUMARICAL DATA

```python
for i in num.columns:
    if data[i].isnull().sum()==0:
        print(f'{i}')
        print(f"Missing Values :{data[i].isnull().sum()}")
        print(f"Mode value :{data[i].mode()}")
        print(f"Unique value :{data[i].unique()}")        
        print(f"No of Unique value :{data[i].nunique()}")
        print('\n')
```

# Distribution of Sales

```python
data['Sales'].plot.hist(bins=20)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()
```

# The histogram illustrates the frequency distribution of sales values, showing that a majority of sales fall within a specific range, with a peak around a certain value.

# How do sales vary over time (e.g., by month)?

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
data.set_index('Order_Date')['Sales'].resample('M').sum().plot()
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.show()
```

# The line chart shows the monthly sales trend over time, revealing fluctuations and potentially identifying sales patterns

# Top product categories by sales

```python
top_categories = data.groupby('Category')['Sales'].sum().nlargest(5)
top_categories.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Top Product Categories by Sales')
ax1 = plt.gca()
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
plt.show()
```

# The bar chart displays the top five product categories by total sales, with data values annotated above each bar

# Does the shipping mode affect delivery times

```python
data['Ship_Date'] = pd.to_datetime(data['Ship_Date'])
data['Delivery_Time'] = (data['Ship_Date'] - data['Order_Date']).dt.days
data.groupby('Ship_Mode')['Delivery_Time'].mean().plot(kind='bar')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Delivery Time (days)')
plt.title('Effect of Shipping Mode on Delivery Time')
plt.show()
```

# The bar chart illustrates that different shipping modes have varying average delivery times, with [Shipping Mode] showing the shortest delivery time and [Shipping Mode] the longest.

# Most popular product sub-categories

```python
popular_subcategories = data['Sub_Category'].value_counts().nlargest(10)
popular_subcategories.plot(kind='barh')
plt.xlabel('Frequency')
plt.ylabel('Sub-Category')
plt.title('Most Popular Product Sub-Categories')
plt.show()
```

# The bar chart shows the top 10 most popular product sub-categories based on frequency, with "Sub-Category" on the y-axis and the number of occurrences on the x-axis.

# sales vary by region

```python
data.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Region')
plt.show()
```

# The pie chart illustrates the distribution of sales by region, with each slice representing the proportion of total sales in a specific region, and the percentage labels highlighting these proportions.

# Top customers by total sales

```python
top_customers = data.groupby('Customer_Name')['Sales'].sum().nlargest(10)
top_customers.plot(kind='bar')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.title('Top Customers by Total Sales')
plt.show()
```

# The bar chart displays the top ten customers ranked by their total sales, with 'Customer Name' on the x-axis and 'Total Sales' on the y-axis.

# What is the average order value over time

```python
data.resample('M', on='Order_Date')['Sales'].mean().plot()
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.title('Average Order Value Over Time')
plt.show()
```

# The line chart shows the average monthly sales over time, indicating a fluctuating trend in the average order value

# How does customer segment impact sales

```python
data.groupby('Segment')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Sales by Customer Segment')
plt.show()
```

# The pie chart illustrates the distribution of sales among different customer segments, with the "Consumer" segment contributing the largest share (50.8%).

```python
shipping_mode_region = data.groupby(['Region', 'Ship_Mode'])['Order_ID'].count().unstack()
shipping_mode_region.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.title('Shipping Modes by Region')
plt.legend(title='Ship Mode')
plt.show()
```

# The stacked bar chart illustrates how different shipping modes are distributed across various regions, showing the distribution of order volumes by shipping mode within each region.
